```python
#IMPORTING THE LIBRARIES
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

#ADJUSTING THE CONFIGURATION OF THE PLOTS TO BE CREATED
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#READING THE DATASET
movies = pd.read_csv('/Users/anoma/Downloads/movies.csv')
```

```python
#CHECKING THE DATA
movies.head()
```

```python
#CHECKING FOR MISSING DATA/ VALUES
for col in movies.columns:
    prt_missing = np.mean(movies[col].isnull())
    print('{} - {}%'.format(col, prt_missing))
```

```python
#REMOVING MISSING DATA/ VALUES
movies = movies.dropna()
```

```python
#RECHECKING FOR MISSING DATA/ VALUES
for col in movies.columns:
    prt_missing = np.mean(movies[col].isnull())
    print('{} - {}%'.format(col, prt_missing))
```

```python
#CHECKING FOR THE DATATYPES 
movies.dtypes
```

```python
#CHANGING DATATYPES OF COLUMMS
movies['votes'] = movies['votes'].astype('int64')

movies['budget'] = movies['budget'].astype('int64')

movies['gross'] = movies['gross'].astype('int64')

movies['runtime'] = movies['runtime'].astype('int64')
```

```python
#RECHECKING IF THE CHANGES WERE EFFECTED
movies.head()
```

```python
#RECHECKING FOR THE DATATYPES 
movies.dtypes
```

```python
# EXTRACTING THE CORRECT YEAR OF RELEASE COLUMN
movies['year'] = movies['released'].str.extract(pat = '([0-9]{4})').astype(int)

#movies.drop('correctreleasedyear',
#           axis='columns', inplace=True)

movies.head()
```

```python
#SORTING THE DATA BY GROSS COLUMN
movies.sort_values(by=['gross'], inplace=False, ascending=False).head()
```

```python
#CHECKING TO SEE THE UNIQUE VALUES IN THE YEAR COLUMN
movies.year.unique()
```

```python
# CHECKING AND DROPPING DUPLICATES IN THE DATA
movies.drop_duplicates()
```

```python
#USING PLOTS VIZ TO SHOW THE CORRELATION BETWEEN BUDGET AND GROSS
#STARTING WITH SCATTERPLOTS
plt.scatter(x=movies['budget'], y=movies['gross'])
plt.title('CORRELATION BETWEEN BUDGET AND GROSS EARNINGS')
plt.xlabel('budget')
plt.ylabel('gross earnings')
plt.show()

#SHOWING WITH SEABORN
sns.regplot(x='budget', y='gross', data=movies, scatter_kws={"color": "black"}, line_kws={"color": "green"})
```

```python
#CHECKING THE CORRELATION BETWEEN NUMERIC COLUMNS USING DIFFERENT METHODS
movies.corr()
#movies.corr(method='pearson')
#movies.corr(method='kendall')
#movies.corr(method='spearman')
```

There is a high correlation between budget and gross.

```python
# VISUALIZING THE CORRELATION MATRIX FOR NUMERIC COLUMNS
correlation_matrix1 = movies.corr(method='pearson')
sns.heatmap (correlation_matrix1, annot=True)
plt.title('Correlation Matrix For Numeric Columns')
plt.show()
```

```python
#SHOWING THE CORRELATION BETWEEN ALL COLUMNS
#First, change the other non-numeric columns to numeric columns
movies_numerized = movies
for col_name in movies_numerized.columns:
    if(movies_numerized[col_name].dtype == 'object'):
        movies_numerized[col_name] = movies_numerized[col_name].astype('category')
        movies_numerized[col_name] = movies_numerized[col_name].cat.codes
movies_numerized.head()
```

```python
movies.head()
```

```python
# VISUALIZING THE CORRELATION MATRIX FOR ALL COLUMNS
correlation_matrix2 = movies_numerized.corr(method='pearson')
sns.heatmap (correlation_matrix2, annot=True)
plt.title('Correlation Matrix For All Columns')
plt.show()
```

```python
#CHECKING THE CORRELATION BETWEEN ALL COLUMNS USING DIFFERENT METHODS
movies_numerized.corr()
#movies_numerized.corr(method='pearson')
#movies_numerized.corr(method='kendall')
#movies_numerized.corr(method='spearman')
```

```python
# FOR ORGANIZATION, UNSTACKING THE OUTPUT ABOVE
matrix2 = movies_numerized.corr()
corr_pairs = matrix2.unstack()

pd.set_option('display.max_rows', None)
sorted_corr = corr_pairs.sort_values()
sorted_corr
```

```python
#SHOWING THE HIGH CORRELATION (>5) BETWEEN ALL COLUMNS
high_corr = sorted_corr[(sorted_corr) > 0.5]
high_corr
```

Votes and budgets have the highest correlation to gross earnings.
